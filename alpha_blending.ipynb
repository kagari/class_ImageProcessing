```python
from PIL import Image
import numpy as np

import matplotlib.pyplot as plt
```

```python
im1 = Image.open("./data/emotions-4637323_1920.jpg")
im1.size
```

```python
im2 = Image.open("./data/anise-2785512_1920.jpg")
im2.size
```

```python
fig, ax = plt.subplots(1,2, figsize=(11,8))
ax[0].imshow(im1)
ax[1].imshow(im2)
plt.show()
```

## 

```python
mask = np.tile(np.linspace(0, 255, 1920), (1280, 1))
mask = Image.fromarray(np.uint8(mask))
```

```python
plt.imshow(mask, cmap="gray")
plt.show()
```

```python
im_alpha_blending = Image.composite(im1, im2, mask)
```

```python
plt.imshow(im_alpha_blending)
plt.show()
```

```python
im_alpha_blending.save('./data/task2/alpha_blending.jpg', quality=300)
```

```python
im_alpha_blending2 = Image.blend(im1, im2, 0.5)
```

```python
plt.imshow(im_alpha_blending2)
plt.show()
```

```python
im_alpha_blending2.save('./data/task2/alpha_blending2.jpg', quality=300)
```
